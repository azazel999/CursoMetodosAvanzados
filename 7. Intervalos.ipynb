### (iii) Implementamos una función a cada operación por separado.

###Suma

```julia
# a' + b' = [a(i) + b(i) , a(s) + b(s) ]
function suma(ai,as,bi,bs)
    set_rounding(Float64, RoundUp) #usamos redondeo hacia arriba
    Intervalo(ai+bi,as+bs)
end
```

```julia
suma(1.0,1.1,3.0,3.1)
```

###Resta

```julia
# a' - b' = [a(i) -  b(s) , a(s) - b(i)]
function resta(ai,as,bi,bs)
    set_rounding(Float64, RoundDown)
    Intervalo(ai-bs,as-bi)
end
```

```julia
resta(1.0,1.1,3.0,3.1)
```

### Multiplicación

```julia
#a'xb'=[min{a(i)b(i),a(i)b(s),a(s)b(i), a(s)b(s)}, max {a(i)b(i), a(i)b(s), a(s)b(i), a(s)b(s)}]
function multiplicacion(ai,as,bi,bs)
    set_rounding(Float64, RoundDown)
    Intervalo(min(ai*bi,ai*bs,as*bi,as*bs),(max(ai*bi,ai*bs,as*bi,as*bs)))
end
```

```julia
multiplicacion(1.0,1.1,3.0,3.1)
```

```julia
#sin los comandos min() y max() obtenemos todos estos valores: 
#(3.0,3.1,3.3000000000000003,3.4100000000000006),(3.0,3.1,3.3000000000000003,3.4100000000000006)
```

###División

```julia
# a' / b' = a' x [1 / b(s), 1 / b(i)], si 0 ∉ b'
function division(ai,as,bi,bs)
    set_rounding(Float64, RoundDown)
    Intervalo(as/bs,ai/bi)
end
```

```julia
division(1.0,1.1,3.0,3.1)
```

####[2] Haz un módulo de Julia llamado Intervalos en un archivo intervalos.jl, que contiene todas las definiciones anteriores.

```julia
module Intervalos

export Intervalo, suma, resta, multiplicacion, division

type Intervalo
    a::Float64 #límite inferior del intervalo
    b::Float64 #límite superior del intervalo
end

# operaciones entre intervalos

#suma

# a' + b' = [a(i) + b(i) , a(s) + b(s) ]
function suma(ai,as,bi,bs)
    set_rounding(Float64, RoundUp) #usamos redondeo hacia arriba
    Intervalo(ai+bi,as+bs)
end

#resta

# a' - b' = [a(i) -  b(s) , a(s) - b(i)]
function resta(ai,as,bi,bs)
    set_rounding(Float64, RoundDown)
    Intervalo(ai-bs,as-bi)
end

#multiplicación

#a'xb'=[min{a(i)b(i),a(i)b(s),a(s)b(i), a(s)b(s)}, max {a(i)b(i), a(i)b(s), a(s)b(i), a(s)b(s)}]
function multiplicacion(ai,as,bi,bs)
    set_rounding(Float64, RoundDown)
    Intervalo(min(ai*bi,ai*bs,as*bi,as*bs),(max(ai*bi,ai*bs,as*bi,as*bs)))
end

# división

# a' / b' = a' x [1 / b(s), 1 / b(i)], si 0 ∉ b'
function division(ai,as,bi,bs)
    set_rounding(Float64, RoundDown)
    Intervalo(as/bs,ai/bi)
end

end
```

```julia
#calcula la suma entre las variables dependientes del polinomio
+(a::ValorPoli,b::ValorPoli) = ValorPoli(a.val,a.pol+b.pol)
```

```julia
#calcula la multiplicación entre las variables dependientes del polinomio
*(a::ValorPoli, b::ValorPoli)=ValorPoli(a.val,a.val*b.val)
```

```julia
#calcula la multiplicación entre un número entero y la variable
*(z::Int64,a::ValorPoli) = ValorPoli(a.val,z*a.val)
```

```julia
#calcula la suma entre un número entero y las componentes del polinomio
+(a::ValorPoli,z::Int64) = ValorPoli(a.val,a.pol+z)
```

```julia
#calcula la resta entre un número entero y las componentes del polinomio
-(a::ValorPoli,z::Int64) = ValorPoli(a.val,a.pol-z)
```
