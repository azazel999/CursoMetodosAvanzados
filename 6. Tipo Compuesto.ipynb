####6.7-i) Define un tipo composito para representar un intervalo de dos números reales.

```julia
type Intervalo
    a::Float64 #límite inferior del intervalo
    x::Float64 #número real contenido dentro del intervalo
    b::Float64 #límite superior del intervalo
end
```

####6.7-ii) ¿Cómo podríamos tener operaciones sensatas sobre los intervalos? La idea básica es que el resultado de la operación sobre dos intervalos contenga los valores posibles resultantes de operar con los miembros de los dos intervalos respectivos.

####Respuesta: En el caso de los intervalos reales se adopta la notación

#### a'=[a(i),a(s)] = {x ∈ R:a(i)≤ x ≤ a(s)}

#### y consideramos el conjunto de todos los intervalos de la recta real como

#### R' = {[a(i),a(s)] : a(i) ≤ a(s) ; a(i), a(s)  ∈ R}

####Sea & una de las operaciones +, -, x, / . Definimos la aritmética sobre los elementos del conjunto real R*, como  

#### a' & b' = {a & b : a ∈ a', b ∈ b'} 

#### con la excepción de que a* / b* no está definida si 0 ∈ b* (b* se define de manera análoga a a*). De esta manera, la artimética para los elementos de R* está dada por

#### a' + b' = [a(i) + b(i) , a(s) + b(s) ]
#### a' - b' = [a(i) -  b(s) , a(s) - b(i)]
#### a' x b' = [min{a(i)b(i), a(i)b(s), a(s)b(i), a(s)b(s)}, max {a(i)b(i), a(i)b(s), a(s)b(i), a(s)b(s)}]
#### a' / b' = a' x [1 / b(s), 1 / b(i)], si 0 ∉ b' 

### 6.7-iii)  Implementa estas operaciones, sin tomar en cuenta cuestiones de redondeo.

#### Sean los intervalos a' = [a(i), a(s)] = [1.0 , 1.1]  y b' = [b(i), b(s)] = [3.0, 3.1]. Entonces,

```julia
type Intervalo
    x::Float64
    y::Float64
end
```

```julia
# a' + b' = [a(i) + b(i) , a(s) + b(s) ]
function suma(ai,as,bi,bs)
    Intervalo(ai+bi,as+bs)
end
```

```julia
suma(1.0,1.1,3.0,3.1)
```

```julia
# a' - b' = [a(i) -  b(s) , a(s) - b(i)]
function resta(ai,as,bi,bs)
    Intervalo(ai-bs,as-bi)
end
```

```julia
resta(1.0,1.1,3.0,3.1)
```

```julia
#a'xb'=[min{a(i)b(i),a(i)b(s),a(s)b(i), a(s)b(s)}, max {a(i)b(i), a(i)b(s), a(s)b(i), a(s)b(s)}]
function multiplicacion(ai,as,bi,bs)
    Intervalo(min(ai*bi,ai*bs,as*bi,as*bs),(max(ai*bi,ai*bs,as*bi,as*bs)))
end
```

```julia
multiplicacion(1.0,1.1,3.0,3.1)
```

```julia
#sin las funciones min y max obtenemos todos estos valores, 
#(3.0,3.1,3.3000000000000003,3.4100000000000006),(3.0,3.1,3.3000000000000003,3.4100000000000006)
# comparando con el resultado anterior, se ve que si toma el mínimo y máximo valor de los 
#intervalos
```

```julia
# a' / b' = a' x [1 / b(s), 1 / b(i)], si 0 ∉ b'
function division(ai,as,bi,bs)
    Intervalo(as/bs,ai/bi)
end
```

```julia
division(1.0,1.1,3.0,3.1)
```
