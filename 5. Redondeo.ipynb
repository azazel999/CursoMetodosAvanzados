####3. El significante es 10011 0011 0011

####0 (signo)| 1111011 (exponente) |  (significante) 10011 0011 0011 0011 0011 001 = 0 1111011  10011 0011 0011 0011 0011 001

#### donde se truncó a 32 bits de precisión.

####Luego, los redondeos hacia arriba y hacia abajo son 

####△(x) = 0 1111011 10011 0011 0011 0011 0011 010

####▽(x) = 0 1111011 10011 0011 0011 0011 0011 001 

####cuya diferencia es
                                                   0.000000000001
                                                   
####Nota: Si comparamos el redondeo hacia arriba obtenido aqui con el resultado dado por "bits", notamos que estos coinciden al final, por lo que la función bits redondea por defecto hacia arriba.

## 5.7 

####La regla x + y = x + y' se cumple para y  y' no se cumple para los números de punto flotante. Por ejemplo: 

#### Sean los números en punto flotante                                                                 
####                                                                     x =  1.10 * 2^1 = 11.0 , y = 1.01 * 2^-1 = 0.101 .   

#### Por consiguiente,                                                   

####                         x + y = 11.0 + 0.101 = 11.101 = 1.1101 * 2^1 (notación científica) = 1.11 * 2^1 (notación de punto flotante). 

#### En notación de punto flotante solo se admiten dos números binarios después del punto, razón por la cual en el ejemplo anterior se omitieron el par de números "01".   

#### Ahora, sea                                                           y' = 1.10 * 2^-1 = 0.110 

#### entonces,                                              
####                                                   x + y' = 11.0 + 0.110 = 11.110 = 1.1110 * 2^1 = 1.11* 2^1  

#### Por lo tanto, se puede apreciar que x + y = x + y' siendo y diferente de y'
